```python
import pandas as pd
import spacy 
import textblob
import nltk
```

```python
PATH = '/content/drive/MyDrive/nlproject/2017-selected-categories-complete-news-2.csv' #the path of the dataset
nlp = spacy.load('en_core_web_sm')
```

```python
df_pakistan = pd.read_csv(PATH) # Reading the data
```

```python
character = {}
for name,article in zip(df_pakistan.author,df_pakistan.body):
  if type(name) == str:
    if not name in character.keys():
      character[name] = []
    character[name].append(article) #Store the article according to the author 
```

```python
import string
authorlist = {}
i = 0
for name in character.keys():
  i = i+1
  if not name in authorlist.keys():
    authorlist[name] = {}
  body_paragraph = character[name]
  print(name)
  for body in body_paragraph:#Obtain the author`s article
    if type(body) == str:
      doc_body = nlp(body)
      for sentences in doc_body.sents:#splite sentence of each article
        sentence = str(sentences)
        doc_sentence = nlp(sentence)
        for ent in doc_sentence.ents:#get the NER, and sentiment anaylise
          if ent.label_ == 'ORG' or ent.label_ == 'PERSON':
            text = ent.text
            blob = textblob.TextBlob(sentence)
            result_sentiment = blob.sentiment.polarity
            if not text in authorlist[name].keys():
              authorlist[name][text] = 0
            authorlist[name][text] += result_sentiment
  if i == 10: # the data is too big so use 10 authors to test
    break
```
